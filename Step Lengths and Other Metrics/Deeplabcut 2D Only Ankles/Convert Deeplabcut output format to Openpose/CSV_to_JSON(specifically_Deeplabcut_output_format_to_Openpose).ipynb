```python
# Remove the first row from the deeplabcut csv, then name the columns of 2nd row distinctly and delete the third row as well
# the unique names help us to identify the components 
import csv
import json
def make_json(csvFilePath, jsonFilePath):
    data = {}
    with open(csvFilePath, encoding='utf-8') as csvf:
        csvReader = csv.DictReader(csvf)
        field = csvReader.fieldnames
        # print(csvReader)
        # i = 0
        for i,rows in enumerate(csvReader):
            # if i < 2:
            #   continue
            # key = rows['bodyparts']
            # data[key] = rows
            # print(rows)
            # rows1 = rows[1:]
            data1 = {"people":[{"pose_keypoints_2d": [float(rows[field[i]]) for i in range(1,len(field))] }]}
            print(data1)
            # DOUBLE CHECK THE FIRST INDEX BY UNCOMMENTING THE NEXT LINE
            # print(rows['scorer'])
            # print(data)
            x = rows['bodyparts']
            # x = rows['scorer']
            # Need to change the zero paddings for more than 999 frames
            if int(x) < 10:
                x1 = '00'+x
            elif int(x) <100:
                x1 = '0'+ x
            else:
                x1 = ''+x
            filename123json= jsonFilePath+ x1 + '.json'
            with open(filename123json, 'w', encoding='utf-8') as jsonf:
                jsonf.write(json.dumps(data1))

            # break
```
